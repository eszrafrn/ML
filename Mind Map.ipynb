## *Introduction*

*Mind Map* adalah sebuah *tool* yang digunakan untuk memetakan ide secara radial dari satu gagasan utama, berguna untuk menyusun ide secara fleksibel dan mempermudah pemahaman. Sedangkan *Concept Map* adalah diagram terstruktur yang menggambarkan hubungan antara berbagai konsep atau topik yang saling terkait.

***Mind Map* VS *Concept Map***
| *Mind Map* | *Concept Map* |
| ----------- | --------- |
| Mengumpulkan berbagai ide | Fokus pada inti informasi yang relevan |
| Fokus pada satu masalah utama | Menghubungkan banyak ide berskala besar |
| Setiap sub-topik dihubungkan menggunakan garis | Relasi antar konsep dihubungkan dengan panah |
| Ide dituangkan sebebas mungkin | Diperlukan analisis mendalam untuk memperoleh solusi|
| Proses pencetusan ide bisa berantakan tapi cepat | Berorientasi pada hasil |
| Lebih cocok diaplikasikan untuk menemukan ide penelitian karena memiliki konsep *free-flowing brainstorming* | Lebih cocok diaplikasikan untuk membuat ide pertanyaan pada suatu penelitian karena dapat membantu untuk menemukan hubungan setiap konsep |
| Strukturnya radial | Strukturnya hirarkis |
